```python
# Question 1: How many movies and TV shows are listed in the dataset?

data = {
    'type': ['Movie', 'TV Show', 'Movie', 'TV Show', 'Movie', 'Movie', 'TV Show', 'Movie']
}
disney_data = pd.DataFrame(data)

# Check if the 'type' column exists in the DataFrame
if 'type' in disney_data.columns:
    type_counts = disney_data['type'].value_counts()
    
    plt.figure(figsize=(6, 6))
    plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'])
    plt.title('Distribution of Movies and TV Shows')
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()
else:
    print("The 'type' column is not present in the DataFrame.")
```

```python
# Plot the top 10 countries by number of titles

data = {
    'country': ['USA', 'UK', 'Canada', 'Australia', 'Germany', 'France', 'Italy', 'Spain', 'Netherlands', 'India', 'Japan'],
    'titles': [150, 120, 100, 90, 85, 80, 75, 70, 65, 60, 55]
}
df = pd.DataFrame(data)

# Count the number of titles per country
country_counts = df.set_index('country')['titles']

# Sort the counts and get the top 10 countries
country_counts = country_counts.sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 4))
sns.barplot(x=country_counts.index, y=country_counts.values, palette='viridis')
plt.title('Top 10 Countries by Number of Titles')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Question 2: Which country has produced the highest number of titles?

data = {
    'country': ['USA', 'UK', 'Canada', 'USA', 'UK', 'USA', 'France', 'Germany', 'UK', 'India', 'Canada', 'France', 'Germany', 'India']
}
disney_data = pd.DataFrame(data)

# Get the top 10 countries by number of titles
country_counts = disney_data['country'].value_counts().head(10)

plt.figure(figsize=(8, 8))
plt.pie(
    country_counts,
    labels=country_counts.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=sns.color_palette('viridis', len(country_counts))
)
plt.title('Top 10 Countries by Number of Titles')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
#Plot the distribution of titles based on their ratings. 

data = {
    'rating': ['PG', 'PG', 'R', 'PG', 'G', 'PG', 'R', 'PG', 'G', 'PG', 'R', 'PG', 'R', 'G']
}
disney_data = pd.DataFrame(data)

# Plot the distribution of titles based on their ratings
plt.figure(figsize=(12, 4))
sns.countplot(data=disney_data, x='rating', palette='viridis')
plt.title('Distribution of Titles by Rating')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Question 3: What is the distribution of titles based on their ratings?

rating_counts = disney_data['rating'].value_counts()
plt.figure(figsize=(7, 7))
plt.pie(rating_counts, labels=rating_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis', len(rating_counts)))
plt.title('Distribution of Titles by Rating')
plt.axis('equal')  
plt.show()
```

```python

# Plot the distribution of titles by rating


plt.figure(figsize=(12, 4))
sns.barplot(x=rating_counts.index, y=rating_counts.values, palette='viridis')
plt.title('Distribution of Titles by Rating')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
